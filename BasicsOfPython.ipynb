```python
# =========================================
# PYTHON BASICS — INTUITION FIRST
# =========================================

# Python reads code from TOP to BOTTOM
# One line at a time

# -----------------------------------------
# 1. PRINT → Python SPEAKS
# -----------------------------------------

print("Hello World")
print("I control the computer")

# -----------------------------------------
# 2. NUMBERS vs TEXT
# Quotes (" ") mean TEXT
# No quotes mean NUMBER
# -----------------------------------------

print(10)        # number
print("10")      # text

# -----------------------------------------
# 3. VARIABLES → Named Boxes
# = means "put into"
# -----------------------------------------

age = 20
name = "Anshu"

print(age)
print(name)

# -----------------------------------------
# 4. PYTHON CAN DO MATH
# -----------------------------------------

print(2 + 3)
print(10 - 4)
print(5 * 2)
print(10 / 2)

# Python remembers numbers using variables
x = 10
y = 3
print(x + y)

# -----------------------------------------
# 5. INPUT → Python LISTENS
# -----------------------------------------

user_name = input("What is your name? ")
print("Hello", user_name)

# -----------------------------------------
# 6. USING INPUT WITH VARIABLES
# -----------------------------------------

user_age = input("Enter your age: ")
print("Your age is", user_age)

# -----------------------------------------
# 7. FINAL MESSAGE
# -----------------------------------------

print("You just wrote your first Python program!")
print("Python listened to you and responded.")
```

```python
# ==================================================
# DECISION MAKING IN PYTHON
# ==================================================

# Until now, Python only followed instructions.
# Now we teach Python HOW TO CHOOSE.

# Python uses:
# if    → check a condition
# elif  → check another condition (only if previous was False)
# else  → do this if nothing above was True


# --------------------------------------------------
# 1. WHAT IS input() REALLY DOING?
# --------------------------------------------------

# input() ALWAYS gives text (string)
# Even if the user types a number

age_text = input("Enter your age: ")
print("What Python received:", age_text)

# age_text is TEXT, not a number


# --------------------------------------------------
# 2. WHAT IS int() AND WHY WE NEED IT
# --------------------------------------------------

# int() converts text into a NUMBER
# Example: "20" → 20

age = int(age_text)
print("After conversion, age is:", age)


# --------------------------------------------------
# 3. WHAT IS if ?
# --------------------------------------------------

# if checks a condition
# If the condition is TRUE → code inside runs
# If FALSE → Python skips it

if age >= 18:
    print("You are an adult.")


# --------------------------------------------------
# 4. WHAT IS else ?
# --------------------------------------------------

# else runs ONLY when if is FALSE

if age >= 18:
    print("You are an adult.")
else:
    print("You are under 18.")


# --------------------------------------------------
# 5. WHAT IS elif ? (else + if)
# --------------------------------------------------

# elif means:
# "If the above condition was false,
# then check this new condition"

marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A+")
elif marks >= 75:
    print("Grade: A")
elif marks >= 60:
    print("Grade: B")
else:
    print("Needs improvement")


# --------------------------------------------------
# 6. IMPORTANT RULE: ORDER MATTERS
# --------------------------------------------------

# Python checks conditions from TOP to BOTTOM
# First TRUE condition runs
# Others are ignored

number = int(input("Enter a number: "))

if number > 0:
    print("Positive number")
elif number == 0:
    print("Zero")
else:
    print("Negative number")


# --------------------------------------------------
# 7. COMPARISON OPERATORS (VERY IMPORTANT)
# --------------------------------------------------

# ==  means "is equal to"
# !=  means "not equal to"
# >   greater than
# <   less than
# >=  greater than or equal to
# <=  less than or equal to

password = input("Enter password: ")

if password == "python123":
    print("Access granted ✅")
else:
    print("Access denied ❌")


# --------------------------------------------------
# 8. WHY INDENTATION (SPACES) MATTER
# --------------------------------------------------

# Indentation tells Python:
# "This code belongs to if / elif / else"

# WRONG (this will break):
# if age >= 18:
# print("Adult")

# CORRECT:
if age >= 18:
    print("Indentation means this line is INSIDE the if")


# --------------------------------------------------
# 9. FINAL UNDERSTANDING
# --------------------------------------------------

print("if → checks a condition")
print("elif → checks another condition")
print("else → runs when all conditions are false")
```
